```python
import tellurium as te
# Enable inline charts
%matplotlib inline
%config InlineBackend.figure_format='retina'
import matplotlib.pyplot as p

# A resuable plotting function for our results
def plot(result, show=True, legend=False):
    if result.dtype.names is None:
        # treat as a regular array
        p.plot(result[:,0], result[:,1:])

    else:
        if len(result.dtype.names) < 1:
            raise Exception('no columns to plot')

        time = result.dtype.names[0]

        for name in result.dtype.names[1:]:
            p.plot(result[time], result[name], label='$' + name + '$')
    
    if legend:
        p.legend(loc = legend)
        
    if show:
        p.show()
```

```python
model = '''
model step1
  J0:    -> S1;  J0_v0
  J1: S1 ->   ;  J1_k3*S1
   # Species initializations:
  S1 = 0
  
  # Variable initializations:
  J0_v0 = 8
  J1_k3 = 1
end
'''

r = te.loadAntimonyModel(model)
results = r.simulate(0, 10, 100)
plot(results, legend=True)
```
